# Theoretical Questions

1. What is unsupervised learning in the context of machine learning ?

Answer : Unsupervised learning is a type of machine learning where the model works with unlabeled data, meaning the data does not contain predefined output labels or target variables. The main objective is to identify hidden patterns, structures, or relationships within the dataset. Common tasks in unsupervised learning include clustering, association rule mining, and dimensionality reduction. For example, customer segmentation in marketing is often performed using clustering algorithms without prior knowledge of customer groups.

2. How does K-Means clustering algorithm work ?

Answer: K-Means clustering is a partition-based clustering algorithm that divides data into K distinct clusters based on similarity. The algorithm starts by randomly selecting K centroids, which act as the center points of clusters. Each data point is then assigned to the nearest centroid using distance measures such as Euclidean distance. After assignment, the centroids are recalculated as the mean of all points within each cluster. This process repeats iteratively until the centroids stabilize and cluster assignments no longer change significantly.

3. Explain the concept of a dendrogram in hierarchical clustering.

Answer: A dendrogram is a tree-like diagram used to visualize the process of hierarchical clustering. It shows how individual data points or clusters are merged step by step based on similarity or distance. The vertical axis represents the distance at which clusters are combined, while the horizontal axis represents the data points. By cutting the dendrogram at a specific height, different numbers of clusters can be obtained. It helps analysts understand the clustering structure and relationships among observations.

4. What is the main difference between K-Means and Hierarchical Clustering?

Answer:The primary difference between K-Means and Hierarchical Clustering lies in how clusters are formed. K-Means requires the user to specify the number of clusters in advance and partitions data into non-overlapping groups using centroid-based optimization. In contrast, Hierarchical Clustering builds a hierarchy of clusters either by merging smaller clusters (agglomerative) or splitting larger clusters (divisive), and it does not always require a predefined number of clusters. Hierarchical clustering also provides a dendrogram for visualization, whereas K-Means does not.

5. What are the advantages of DBSCAN over K-Means?

Answer: DBSCAN offers several advantages over K-Means, especially for complex datasets. It can identify clusters of arbitrary shapes rather than only spherical clusters. DBSCAN is also effective at detecting noise and outliers, which K-Means usually forces into clusters. Another advantage is that DBSCAN does not require specifying the number of clusters beforehand. Additionally, it performs better when clusters vary in size and density, although parameter tuning is important for optimal results.

6. When would you use Silhouette Score in clustering?

Answer: Silhouette Score is used to evaluate the quality of clustering results by measuring how well data points fit within their assigned clusters compared to other clusters. It is especially useful when selecting the optimal number of clusters in algorithms like K-Means. A higher Silhouette Score indicates that clusters are well-separated and compact, while lower scores suggest overlapping or poorly defined clusters. It provides a more meaningful evaluation than just measuring cluster compactness alone.

7. What are the limitations of Hierarchical Clustering?

Answer: Hierarchical Clustering has several limitations, particularly with large datasets. It is computationally expensive because it calculates distances between all data points, making it slower than algorithms like K-Means. Once clusters are merged or split, the process cannot be reversed, which may lead to suboptimal clustering. The algorithm is also sensitive to noise and outliers, and selecting the appropriate linkage method can significantly affect results.

8. Why is feature scaling important in clustering algorithms like K-Means?

Answer: Feature scaling is important in clustering because algorithms like K-Means rely on distance calculations to group data points. If features have different scales, variables with larger numerical ranges can dominate the distance measurement and bias the clustering results. Scaling methods such as normalization or standardization ensure that all features contribute equally to the clustering process, leading to more accurate and meaningful clusters.

9. How does DBSCAN identify noise points?

Answer: DBSCAN identifies noise points by analyzing the density of neighboring data points. A point is considered noise if it does not have enough nearby points within a specified radius called epsilon (ε). Specifically, if a point has fewer neighboring points than the minimum required threshold (MinPts), and it is not reachable from a core point, it is labeled as noise or an outlier. This capability makes DBSCAN effective for anomaly detection.

10. Define inertia in the context of K-Means.

Answer: Inertia in K-Means refers to the sum of squared distances between each data point and its nearest cluster centroid. It measures how compact and internally cohesive the clusters are. Lower inertia values indicate tighter clusters and better clustering performance. However, inertia alone is not always sufficient for evaluating clustering quality because it decreases as the number of clusters increases.

11. What is the elbow method in K-Means clustering?

Answer: The elbow method is a technique used to determine the optimal number of clusters in K-Means clustering. It involves plotting the number of clusters against the corresponding inertia values. As the number of clusters increases, inertia decreases. The optimal cluster count is typically identified at the “elbow point,” where the rate of decrease in inertia sharply slows down. This point represents a balance between cluster compactness and model simplicity.

12. Describe the concept of "density" in DBSCAN.

Answer: In DBSCAN, density refers to the concentration of data points within a specific neighborhood radius. A region is considered dense if it contains at least a minimum number of points (MinPts) within the epsilon radius. Clusters are formed by connecting dense regions together, while sparse regions are treated as noise. This density-based approach allows DBSCAN to detect irregularly shaped clusters effectively.

13. Can hierarchical clustering be used on categorical data?

Answer: Yes, hierarchical clustering can be used on categorical data if an appropriate similarity or distance measure is applied. Traditional distance measures like Euclidean distance are not suitable for categorical variables. Instead, measures such as Hamming distance, Jaccard similarity, or Gower distance are commonly used. This flexibility allows hierarchical clustering to work with both numerical and categorical datasets.

14. What does a negative Silhouette Score indicate?

Answer: A negative Silhouette Score indicates that a data point may have been assigned to the wrong cluster. It means the point is closer to points in another cluster than to points within its own cluster. Negative scores suggest poor clustering performance, overlapping clusters, or incorrect cluster assignments. Ideally, clustering models should aim for Silhouette Scores close to +1.

15. Explain the term "linkage criteria" in hierarchical clustering.

Answer: Linkage criteria in hierarchical clustering determine how the distance between two clusters is calculated during the merging process. Common linkage methods include single linkage, complete linkage, average linkage, and Ward’s linkage. Single linkage uses the minimum distance between points in two clusters, while complete linkage uses the maximum distance. The choice of linkage criterion significantly affects the shape and structure of the resulting clusters.

16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

Answer: K-Means assumes that clusters are spherical, equally sized, and have similar densities. When clusters vary significantly in size or density, K-Means may incorrectly assign points because it relies on centroid distance alone. Dense clusters can dominate sparse clusters, and irregularly shaped clusters may not be properly separated. As a result, the algorithm may produce inaccurate or misleading clustering outcomes.

17. What are the core parameters in DBSCAN, and how do they influence clustering?

Answer: The two core parameters in DBSCAN are epsilon (ε) and MinPts. Epsilon defines the radius within which neighboring points are searched, while MinPts specifies the minimum number of points required to form a dense region. A small epsilon may result in many points being labeled as noise, whereas a large epsilon can merge distinct clusters together. Similarly, higher MinPts values require denser regions to form clusters. Proper parameter selection is essential for effective clustering.

18. How does K-Means++ improve upon standard K-Means initialization?

Answer: K-Means++ improves standard K-Means by selecting initial centroids more strategically rather than randomly. The first centroid is chosen randomly, while subsequent centroids are selected with higher probability from points farther away from existing centroids. This approach spreads centroids across the dataset and reduces the chances of poor initialization. As a result, K-Means++ often converges faster and produces more accurate clustering results.

19. What is agglomerative clustering?

Answer: Agglomerative clustering is a bottom-up hierarchical clustering technique where each data point initially forms its own cluster. The algorithm then repeatedly merges the two most similar clusters until all points belong to a single cluster or a stopping condition is reached. Similarity between clusters is determined using linkage criteria such as single, complete, or average linkage. The process is commonly visualized using a dendrogram.

20. What makes Silhouette Score a better metric than just inertia for model evaluation?

Answer: Silhouette Score is considered better than inertia because it evaluates both cluster compactness and separation simultaneously. Inertia only measures how close points are to their cluster centroids and always decreases as more clusters are added, which can lead to overfitting. In contrast, Silhouette Score assesses whether points are appropriately assigned to clusters by comparing intra-cluster similarity and inter-cluster separation. This provides a more balanced and reliable evaluation of clustering quality.

```python
"21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot"

# Import required libraries
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate synthetic data with 4 centers
X, y = make_blobs(
    n_samples=300,
    centers=4,
    cluster_std=1.0,
    random_state=42
)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)

# Get cluster labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Visualize the clusters
plt.figure(figsize=(8, 6))

# Scatter plot of data points
plt.scatter(X[:, 0], X[:, 1], c=labels)

# Plot centroids
plt.scatter(
    centroids[:, 0],
    centroids[:, 1],
    s=300,
    marker='X'
)

# Add labels and title
plt.title("K-Means Clustering with 4 Centers")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

# Show plot
plt.show()
```

```python
"29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart"

# Import required libraries
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Generate synthetic data
X, y = make_blobs(
    n_samples=300,
    centers=4,
    cluster_std=1.0,
    random_state=42
)

# Range of k values
k_values = range(2, 6)

# Store silhouette scores
silhouette_scores = []

# Compute silhouette score for each k
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)

    score = silhouette_score(X, labels)
    silhouette_scores.append(score)

# Plot bar chart
plt.figure(figsize=(8, 6))

plt.bar(k_values, silhouette_scores)

# Add labels and title
plt.title("Silhouette Scores for Different K Values")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")

# Show plot
plt.show()
```

```python
"31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries"

# Import required libraries
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Generate synthetic data with overlapping clusters
X, y = make_blobs(
    n_samples=400,
    centers=3,
    cluster_std=2.5,
    random_state=42
)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Create mesh grid for decision boundaries
h = 0.02  # Step size

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(
    np.arange(x_min, x_max, h),
    np.arange(y_min, y_max, h)
)

# Predict cluster labels for each grid point
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundaries
plt.figure(figsize=(10, 7))

plt.contourf(xx, yy, Z, alpha=0.3)

# Plot original data points
plt.scatter(
    X[:, 0],
    X[:, 1],
    c=kmeans.labels_
)

# Plot centroids
plt.scatter(
    kmeans.cluster_centers_[:, 0],
    kmeans.cluster_centers_[:, 1],
    s=300,
    marker='X'
)

# Add labels and title
plt.title("K-Means Clustering with Decision Boundaries")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

# Show plot
plt.show()
```

```python
"40. Load the Digits dataset, apply PCA to reduce to 3 components, then use KMeans and visualize with a 3D scatter plot."

# Import required libraries
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the Digits dataset
digits = load_digits()
X = digits.data

# Reduce dimensions to 3 components using PCA
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X)

# Apply K-Means clustering
kmeans = KMeans(
    n_clusters=10,
    random_state=42
)

labels = kmeans.fit_predict(X_pca)

# Create 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot clustered data points
ax.scatter(
    X_pca[:, 0],
    X_pca[:, 1],
    X_pca[:, 2],
    c=labels
)

# Plot cluster centers
centers = kmeans.cluster_centers_

ax.scatter(
    centers[:, 0],
    centers[:, 1],
    centers[:, 2],
    s=300,
    marker='X',
    label='Centroids'
)

# Add labels and title
ax.set_title("3D PCA Visualization of K-Means Clustering on Digits Dataset")
ax.set_xlabel("Principal Component 1")
ax.set_ylabel("Principal Component 2")
ax.set_zlabel("Principal Component 3")

ax.legend()

# Show plot
plt.show()
```

```python
"44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering"

# Import required libraries
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.pyplot as plt
import numpy as np

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-Means clustering
kmeans = KMeans(
    n_clusters=3,
    random_state=42
)

labels = kmeans.fit_predict(X_scaled)

# Compute silhouette scores for each sample
sample_silhouette_values = silhouette_samples(X_scaled, labels)

# Compute average silhouette score
avg_score = silhouette_score(X_scaled, labels)

# Create silhouette plot
plt.figure(figsize=(10, 6))

y_lower = 10

for i in range(3):
    # Get silhouette scores for samples in cluster i
    ith_cluster_values = sample_silhouette_values[labels == i]

    # Sort the values
    ith_cluster_values.sort()

    cluster_size = ith_cluster_values.shape[0]
    y_upper = y_lower + cluster_size

    # Fill silhouette area
    plt.fill_betweenx(
        np.arange(y_lower, y_upper),
        0,
        ith_cluster_values,
        alpha=0.7
    )

    # Label cluster number
    plt.text(-0.05, y_lower + 0.5 * cluster_size, str(i))

    y_lower = y_upper + 10

# Draw average silhouette score line
plt.axvline(
    x=avg_score,
    linestyle='--',
    label=f'Average Score = {avg_score:.2f}'
)

# Add labels and title
plt.title("Silhouette Plot for K-Means Clustering on Iris Dataset")
plt.xlabel("Silhouette Coefficient Values")
plt.ylabel("Cluster")

plt.legend()

# Show plot
plt.show()
```

```python
"46. Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4 features)"

# Import required libraries
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Wine dataset
wine = load_wine()

# Create DataFrame with first 4 features
df = pd.DataFrame(
    wine.data[:, :4],
    columns=wine.feature_names[:4]
)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

# Apply K-Means clustering
kmeans = KMeans(
    n_clusters=3,
    random_state=42
)

# Predict cluster labels
df['Cluster'] = kmeans.fit_predict(X_scaled)

# Create seaborn pairplot
sns.pairplot(
    df,
    vars=wine.feature_names[:4],
    hue='Cluster'
)

# Show plot
plt.show()
```

```python
"48. Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters."

# Import required libraries
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Load the Digits dataset
digits = load_digits()
X = digits.data

# Reduce dimensions to 2D using t-SNE
tsne = TSNE(
    n_components=2,
    random_state=42,
    perplexity=30
)

X_tsne = tsne.fit_transform(X)

# Apply Agglomerative Clustering
agg_clustering = AgglomerativeClustering(
    n_clusters=10,
    linkage='ward'
)

labels = agg_clustering.fit_predict(X_tsne)

# Plot the clustering result
plt.figure(figsize=(10, 7))

plt.scatter(
    X_tsne[:, 0],
    X_tsne[:, 1],
    c=labels
)

# Add labels and title
plt.title("Agglomerative Clustering on Digits Dataset (t-SNE Reduced)")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")

# Show plot
plt.show()
```
